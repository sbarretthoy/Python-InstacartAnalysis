```python
# define a function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# apply the function
df['price_range'] = df.apply(price_label, axis=1)
```

1. create price_label and busiest_day column

```python
# create price_label for low, mid, and high range product using loc function 
ords_prods_merge.loc[ords_prods_merge['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
ords_prods_merge.loc[(ords_prods_merge['prices'] <= 15) & (ords_prods_merge['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
ords_prods_merge.loc[ords_prods_merge['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
ords_prods_merge['price_range_loc'].value_counts(dropna = False)
```

```python
#check orders_day_of_week frequencies
ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

```python
# create for-loop for busiest day
result = []

for value in ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
# create column for result in ords_prods_merge dataframe
ords_prods_merge['busiest_day'] = result
```

```python
ords_prods_merge['busiest_day'].value_counts(dropna = False)
```

2. change busiest_day to busiest_days

```python
# create for-loop for busiest days
result = []

for value in ords_prods_merge["orders_day_of_week"]:
  if value == 0 or value == 1:
    result.append("Busiest days")
  elif value == 3 or value == 4:
    result.append("Least busy days")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
# create column for busiest_days in ords_prods_merge dataframe
ords_prods_merge['busiest_days'] = result
```

```python
ords_prods_merge['busiest_days'].value_counts(dropna = False)
```

4. identify busiest hours of day by “Most orders,” “Average orders,” and “Fewest orders.”

```python
# check order_hour_of_day frequencies
ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
# create for-loop for busiest hour of day separating the hours into equal thirds
result = []

for value in ords_prods_merge["order_hour_of_day"]:
  if value in [9, 10, 11, 12, 13, 14, 15, 16]:
    result.append("Most orders")
  elif value in [7, 8, 17, 18, 19, 20, 21, 22]:
    result.append("Average orders")
  else:
    result.append("Fewest orders")
```

```python
result
```

```python
# create column for busiest_period_of_day in ords_prods_merge dataframe
ords_prods_merge['busiest_period_of_day'] = result
```
