## Part 1.6 Visualizations of distribution of profiles

## Part 1.9 Charts to represent results

```python
# create chart of income profile and department
instacart_final_income_dep = instacart_final.groupby(['Income Profile','departments']).size().unstack()
```

```python
# sort by ascending
instacart_final_income_dep = instacart_final_income_dep.loc[instacart_final_income_dep.sum(axis = 1).sort_values(ascending = True).index]
```

```python
instacart_final_income_dep
```

```python
# create chart of age profile and department
instacart_final_age_dep = instacart_final.groupby(['Age Profile','departments']).size().unstack()
```

```python
# sort by ascending
instacart_final_age_dep = instacart_final_age_dep.loc[instacart_final_age_dep.sum(axis = 1).sort_values(ascending = True).index]
```
