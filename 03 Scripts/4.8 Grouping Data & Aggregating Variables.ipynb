2. Aggregate entire dataframe

```python
ords_prods_merge.groupby('department_id').agg({'order_number': ['mean']})
```

3. The mean for each department id has changed, the example only looked at a portion of the dataset which altered the mean. Now with all of the data being included we can see more accurate numbers.

4. create loyalty flag

```python
# use transform to find max order number for each user
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# check output
ords_prods_merge.head()
```

```python
# use loc to create loyalty flag based on predefined numbers
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# check counts for loyalty flag
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# check output
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

5. check basic spending habits of customers

```python
ords_prods_merge.groupby('loyalty_flag'). agg({'prices' : ['mean', 'min', 'max']})
```

6. create a spending flag based on spending habits

```python
# use transorm to find average price of products purchased by each user
ords_prods_merge['avg_order'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# create spending falg for each user using predefined terms
ords_prods_merge.loc[ords_prods_merge['avg_order'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_order'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# check counts for spending flag
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# check output
ords_prods_merge[['user_id', 'avg_order', 'spending_flag']].head(60)
```

7. determine frequent versus non-frequent customers

```python
# find median of each users days since prior order using groupby
ords_prods_merge['median_order_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge[['user_id', 'days_since_prior_order', 'median_order_frequency']].head(20)
```

```python
# use loc to create frequency flag based on predefined numbers
ords_prods_merge.loc[ords_prods_merge['median_order_frequency'] > 20, 'order_frequency'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_order_frequency'] <= 20) & (ords_prods_merge['median_order_frequency'] > 10), 'order_frequency'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_order_frequency'] <= 10, 'order_frequency'] = 'Frequent customer'
```

```python
# check counts of ew order frequency flag
ords_prods_merge['order_frequency'].value_counts(dropna = False)
```
